```python
from bs4 import BeautifulSoup
import requests
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
import html
from selenium.webdriver.common.by import By
import time
import pandas as pd
import numpy as np
from selenium.webdriver.common.action_chains import ActionChains
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'
}
page = requests.get("https://www.rtbf.be/en-continu/sport", headers=headers)

if page.status_code == 200: 
    print("nice")

soup = BeautifulSoup(page.text, 'html.parser')
#page.text
```

```python
#quotes = soup.find_all('li', class_='js-uptime')

quotes = soup.find_all('div', class_='flex flex-col md:px-24 px-20 py-16 w-3/5 bg-card-light dark:bg-card-dark md:w-3/4')
```

```python
quotes = soup.find_all('div', class_='flex flex-col md:px-24 px-20 py-16 w-3/5 bg-card-light dark:bg-card-dark md:w-3/4')

titles = []  # Initialize an empty list to store titles
more_details = []
for quote in quotes:
    # Select the third-level nested div within each 'flex-col' div
    nested_div = quote.select('div')

    for nest in nested_div:
        paragraph = nest.select('p') # Only first match
        headers  = nest.select('header') # Only first match
        titles.append(headers[0].text)
        more_details.append(paragraph[0].text)
        print(nest.attrs)


#data = np.array([titles, more_details])  
data = np.array(list(zip(titles, more_details)))

data
```

```python


driver = webdriver.Chrome()
# Example of navigating to a page
driver.get("https://www.rtbf.be/en-continu/info")
time.sleep(3)

start_time = time.time()

titles=[]
links=[]
timeArticle=[]
buttonCookies = driver.find_element(By.CSS_SELECTOR,"button#didomi-notice-agree-button")
buttonCookies.click()
time.sleep(2)

while time.time() - start_time < 6:
    
    button = driver.find_element("xpath", "//button[span[text()='Charger 20 articles en plus']]")
    ActionChains(driver).move_to_element(button).perform()
    button.click()
    time.sleep(1.5)

#après avoir chargé la page, on commence à scrapper :>)
elements = driver.find_elements(By.CSS_SELECTOR, "div.flex-1")

for element in elements : 

    content = element.find_element(By.CSS_SELECTOR, "header h3 a")
    article_link = element.find_element(By.CSS_SELECTOR, "header h3 a").get_attribute("href")  # This will print the text within the <a> tag
    link_html = content.get_attribute("innerHTML")
    article_title = html.unescape(link_html)  # This decodes HTML entities like &nbsp;
    titles.append(article_title)
    links.append(article_link)

for link in links : 
    driver.get(link)
    timeTag= driver.find_element(By.CSS_SELECTOR, "span.card-meta.text-content-900.dark\\:text-content-400")
    time.sleep(0.5)
    timeArticle.append(timeTag.text)

#pour stocker les données sur forme de dataframe
data =  np.array(list(zip(titles,links,timeArticle)))
datframe = pd.DataFrame(data)
driver.quit()
datframe
```
